```python
import pandas as pd
from datasets import load_dataset
```

```python
# stance_feminist

dataset_stance_abortion = []
dataset = load_dataset("tweet_eval", "stance_feminist")
dataset_stance_abortion.append(dataset)

stance_abortion_corpus=[]
stance_abortion_corpus.extend(dataset_stance_abortion[0]['train'])
  
stance_abortion_label=[]
stance_abortion_text=[]
for i in range(len(stance_abortion_corpus)):
  
  stance_abortion_label.append(stance_abortion_corpus[i]['label'])
  stance_abortion_text.append(stance_abortion_corpus[i]['text'])


df_stance_abortion_label=pd.DataFrame(stance_abortion_label,columns=['label'])
df_stance_abortion_text=pd.DataFrame(stance_abortion_text,columns=['text'])

df_stance_abortion=pd.concat([df_stance_abortion_text,df_stance_abortion_label],axis=1)


df_stance_abortion_favor=df_stance_abortion[df_stance_abortion['label']==2]
df_stance_abortion_against=df_stance_abortion[df_stance_abortion['label']==1]

df_stance_abortion_favor_asc=df_stance_abortion_favor['text'].sort_values(ascending=True)
df_stance_abortion_favor_dsc=df_stance_abortion_favor['text'].sort_values(ascending=False)

df_stance_abortion_all = pd.DataFrame(zip(df_stance_abortion_favor_asc,df_stance_abortion_favor_dsc,df_stance_abortion_against['text']),columns=['sent0','sent1','hard_neg'])

df_stance_abortion_1 = df_stance_abortion_all.drop(df_stance_abortion_all[df_stance_abortion_all['sent0']==df_stance_abortion_all['sent1']].index)


df_stance_abortion_1.to_csv('tweeteval_stance_feminist',

                 sep=',',

                 na_rep='NaN', 

                 columns = ['sent0','sent1','hard_neg'], # columns to write

                 index = False) # do not write index
```
